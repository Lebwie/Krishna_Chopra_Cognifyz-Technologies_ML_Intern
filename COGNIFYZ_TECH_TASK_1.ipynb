# Loading the DataSet

## DataPreparation

### Data Cleaning

```python
df.isnull().sum()
```

Filling the null values to "***Unknown***"

## Data Separation

```python

x = df.drop('Aggregate rating', axis=1)
y = df['Aggregate rating']
```

## Data Spliting

## X and y Training datset

```python
 from sklearn.model_selection import train_test_split

 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_test
```

# Model Building


### Lable Encoding

```python
import pandas as pd

# Apply one-hot encoding to all object columns in both train and test sets
x_train = pd.get_dummies(x_train, columns=x_train.select_dtypes(include='object').columns)
x_test = pd.get_dummies(x_test, columns=x_test.select_dtypes(include='object').columns)

# Align columns - crucial for consistent feature sets
x_train, x_test = x_train.align(x_test, join='outer', axis=1, fill_value=0)
```

## **Linera Regaression**

### Model Fitting

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train )
```

### **Applying the model to make prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

**Result**

```python
print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### **Evaluation**

### Model Performance

```python
from sklearn.metrics import r2_score , mean_squared_error

y_train_lr_mse = mean_squared_error(y_train, y_lr_train_pred)
y_train_lr_r2 = r2_score(y_train, y_lr_train_pred)

y_test_lr_mse = mean_squared_error(y_test, y_lr_test_pred)
y_test_lr_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE TRAIN:', y_train_lr_mse)
print('LR MSE TEST:', y_test_lr_mse)
print('LR R2 TRAIN:', y_train_lr_r2)
print('LR R2 TEST:', y_test_lr_r2)
```

### Results Table

```python
lr_results = pd.DataFrame(['LinearRegression', y_train_lr_mse, y_test_lr_mse, y_train_lr_r2 , y_test_lr_r2]).transpose()
lr_results.columns = ['Meathod', 'Training mse', 'Testing mse', 'Training r2' , 'Testing r2']
```

```python
lr_results
```
